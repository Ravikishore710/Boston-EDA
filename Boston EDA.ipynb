```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
```

```python
df = pd.read_csv('HousingData.csv')

print("First 5 rows of the dataset:")
print(df.head())

print("\nDataset Information:")
df.info()

print("\nSummary Statistics:")
print(df.describe())

print("\nMissing values per column:")
print(df.isnull().sum())
```

```python

# 1. Distribution of the target variable (MEDV)
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='MEDV', bins=30, kde=True)
plt.title('Distribution of Median Home Value (MEDV)')
plt.xlabel('Median Home Value ($1000s)')
plt.ylabel('Frequency')
plt.show()

# 2. Relationship between Crime Rate (CRIM) and Median Home Value (MEDV)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='CRIM', y='MEDV')
plt.title('Crime Rate vs. Median Home Value')
plt.xlabel('Per Capita Crime Rate')
plt.ylabel('Median Home Value ($1000s)')
plt.show()

# 3. Relationship between Number of Rooms (RM) and Median Home Value (MEDV)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='RM', y='MEDV')
plt.title('Number of Rooms vs. Median Home Value')
plt.xlabel('Average Number of Rooms')
plt.ylabel('Median Home Value ($1000s)')
plt.show()
```

```python
# 4. Relationship between LSTAT and MEDV
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='LSTAT', y='MEDV')
plt.title('% Lower Status of Population vs. Median Home Value')
plt.xlabel('% Lower Status of Population')
plt.ylabel('Median Home Value ($1000s)')
plt.show()

# 5. Relationship between PTRATIO and MEDV (Scatter plot)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='PTRATIO', y='MEDV')
plt.title('Pupil-Teacher Ratio vs. Median Home Value')
plt.xlabel('Pupil-Teacher Ratio')
plt.ylabel('Median Home Value ($1000s)')
plt.show()

# 6. Box plot of MEDV by CHAS (Charles River dummy variable)
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='CHAS', y='MEDV')
plt.title('Median Home Value by Charles River Proximity')
plt.xlabel('Charles River Proximity (0 = No, 1 = Yes)')
plt.ylabel('Median Home Value ($1000s)')
plt.xticks([0, 1], ['Not bounds River', 'Bounds River'])
plt.show()
```

```python
# 7. Distribution of TAX (Property Tax Rate)
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='TAX', bins=30, kde=True)
plt.title('Distribution of Property Tax Rate')
plt.xlabel('Property Tax Rate')
plt.ylabel('Frequency')
plt.show()

# 8. Distribution of AGE (Proportion of older units)
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='AGE', bins=30, kde=True)
plt.title('Distribution of Proportion of Units Built Prior to 1940')
plt.xlabel('Proportion of Units Built Prior to 1940')
plt.ylabel('Frequency')
plt.show()

# 9. Correlation matrix (Heatmap)
plt.figure(figsize=(12, 10))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Boston Housing Features')
plt.show()

# 10. Pair plot for a subset of features (can be time-consuming for many features)
selected_features = ['CRIM', 'RM', 'LSTAT', 'MEDV']
sns.pairplot(df[selected_features])
plt.suptitle('Pair Plot of Selected Boston Housing Features', y=1.02)
plt.show()
```
